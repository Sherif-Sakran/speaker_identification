```python
def replicate_directory_structure(src_dir, dst_dir):
    """
    Replicates the directory structure from src_dir into dst_dir.
    """
    for root, dirs, _ in os.walk(src_dir):
        for dir_name in dirs:
            src_path = os.path.join(root, dir_name)
            relative_path = os.path.relpath(src_path, src_dir)
            dst_path = os.path.join(dst_dir, relative_path)
            os.makedirs(dst_path, exist_ok=True)
```

```python
def process_audio_to_mfcc_csv(src_dir, dst_dir, n_mfcc=22):
    """
    Processes all .wav files in src_dir, computes MFCC features,
    and saves them as .csv files in dst_dir.
    """
    for root, _, files in os.walk(src_dir):
        for file_name in files:
            if file_name.lower().endswith('.wav'):
                # Construct full paths
                src_file_path = os.path.join(root, file_name)
                relative_path = os.path.relpath(root, src_dir)
                dst_file_dir = os.path.join(dst_dir, relative_path)
                os.makedirs(dst_file_dir, exist_ok=True)
                
                # Compute MFCCs
                try:
                    signal, sr = librosa.load(src_file_path)
                    mfccs = librosa.feature.mfcc(y=signal, n_mfcc=n_mfcc, sr=sr)
                    delta_mfccs = librosa.feature.delta(mfccs)
                    delta2_mfccs = librosa.feature.delta(mfccs, order=2)
                    mfcc_delta_dd = np.concatenate((mfccs, delta_mfccs, delta2_mfccs)).T
                    
                    # Save as CSV
                    base_name = os.path.splitext(file_name)[0]
                    dst_file_path = os.path.join(dst_file_dir, f"{base_name}.csv")
                    pd.DataFrame(mfcc_delta_dd).to_csv(dst_file_path, index=False)
                    print(f"Processed and saved: {dst_file_path}")
                except Exception as e:
                    print(f"Error processing {src_file_path}: {e}")
```

```python
src_directory = "../Datasets"
n_mfcc = 22
dst_directory = f"{src_directory}_mfcc_{n_mfcc}"

# Step 1: Replicate directory structure
replicate_directory_structure(src_directory, dst_directory)

# Step 2: Process audio files and generate MFCC CSVs
process_audio_to_mfcc_csv(src_directory, dst_directory, n_mfcc=n_mfcc)
```
