```python
import noisereduce as nr
import os
from scipy.io.wavfile import read, write
```

```python
training = False
if training:
    dir = '3-wav_training_noise_preprocessed'
else:
    dir = '3-wav_testing_noise_preprocessed'
```

```python
import glob

wav_files = glob.glob(dir + '/**/*.wav', recursive=True)
wav_files
```

```python
for path in wav_files:
    # load data
    rate, data = read(path)
    # perform noise reduction
    reduced_noise = nr.reduce_noise(y=data, sr=rate)
    write(path, rate, reduced_noise)
```
